```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, poisson

# Parameters for the Gamma distribution
shape_param = 2
scale_param = 1

# Simulate 100 lambda values from the Gamma distribution
np.random.seed(20240808)
lambda_values = np.random.gamma(shape_param, scale_param, 500)
# Plot the observed data as a histogram
plt.figure(figsize=(10, 6))

aggregated_pmf_2_1 = np.zeros(16)

# Overlay the expected Poisson distribution
x_2_1 = np.arange(0, 16)
for lambda_val in lambda_values:
    aggregated_pmf_2_1 += poisson.pmf(x_2_1, lambda_val)
    # plt.plot(x, poisson_pmf, 'o-', color='#00CED1', alpha=0.02, markersize=8)

# Normalize the aggregated PMF
aggregated_pmf_2_1 /= len(lambda_values)

plt.plot(x_2_1, aggregated_pmf_2_1, 'o-', color='#8A2BE2', markersize=8)

# Set x-axis labels for every element
plt.xticks(np.arange(0, 16))

plt.title('Simulated Poisson(λ) Distribution with λ from Gamma(2,1) Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Parameters for the Gamma distribution
shape_param = 7
scale_param = 0.5

# Simulate 100 lambda values from the Gamma distribution
np.random.seed(20240808)
lambda_values_7_2 = np.random.gamma(shape_param, scale_param, 500)
# Plot the observed data as a histogram
plt.figure(figsize=(10, 6))

aggregated_pmf_7_2 = np.zeros(16)

# Overlay the expected Poisson distribution
x_7_2 = np.arange(0, 16)
for lambda_val in lambda_values_7_2:
    aggregated_pmf_7_2 += poisson.pmf(x_7_2, lambda_val)
    # plt.plot(x, poisson_pmf, 'o-', color='#00CED1', alpha=0.02, markersize=8)

# Normalize the aggregated PMF
aggregated_pmf_7_2 /= len(lambda_values)

plt.plot(x_7_2, aggregated_pmf_7_2, 'o-', color='#00CED1', markersize=8)


# Set x-axis labels for every element
plt.xticks(np.arange(0, 16))

plt.title('Simulated Poisson(λ) Distribution with λ from Gamma(7,2) Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the aggregated PMFs side by side
plt.figure(figsize=(20, 6))

# First subplot
plt.subplot(1, 2, 1)
plt.plot(x_2_1, aggregated_pmf_2_1, 'o-', color='#8A2BE2', markersize=8)
plt.xticks(np.arange(0, 16))
plt.title('Prior Poisson(λ) Distribution with λ from Gamma(2,1) Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Second subplot
plt.subplot(1, 2, 2)
plt.plot(x_7_2, aggregated_pmf_7_2, 'o-', color='#00CED1', markersize=8)
plt.xticks(np.arange(0, 16))
plt.title('Posterior Poisson(λ) Distribution with λ from Gamma(7,2) Distribution')
plt.xlabel('Value')
```
